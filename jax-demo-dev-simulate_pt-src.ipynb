```python
lens_prior = tfd.JointDistributionSequential(
    [
        tfd.JointDistributionNamed(
            dict(
                theta_E=tfd.LogNormal(jnp.log(1.2), 0.25),
                center_x=tfd.Normal(0, 0.01),
                center_y=tfd.Normal(0, 0.01),
            )
        ),
    ]
)
lens_light_prior = tfd.JointDistributionSequential(
    [
        tfd.JointDistributionNamed(
            dict(
                R_sersic=tfd.LogNormal(jnp.log(1.0), 0.2),
                n_sersic=tfd.Uniform(3, 5),
                center_x=tfd.Normal(0, 0.1),
                center_y=tfd.Normal(0, 0.1),
                Ie=tfd.LogNormal(jnp.log(1.5), 0.2),
            )
        )
    ]
)

source_light_prior = tfd.JointDistributionSequential(
    [
        tfd.JointDistributionNamed(
            dict(
                center_x=tfd.Normal(0.02, 0.1),
                center_y=tfd.Normal(0, 0.1),
                Ie=tfd.LogNormal(jnp.log(200.0), 0.5),
            )
        )
    ]
)

prior = tfd.JointDistributionSequential(
    [lens_prior, lens_light_prior, source_light_prior]
)

path = os.path.expanduser('~') + '/HST-modeling/SN-Zwicky/'
kernel = np.load(path + 'F475W-epsf.npy').astype(np.float32)
```

```python
from copy import copy
delta_pix, num_pix = 0.02, 61
ctr = num_pix//2
supersample = 1

theta_E = 0.18
truth = [[{'theta_E': theta_E, 'center_x': 0.0, 'center_y': 0.0}], 
         [{'R_sersic': 0.8, 'n_sersic': 4.0, 'center_x': 0.0, 'center_y': 0.0, 'Ie': 0.0001}], 
         [{'center_x': 0.03, 'center_y': 0.0, 'Ie': 300000}]]

src_truth = copy(truth); src_truth[0][0]['theta_E'] = 1e-6; src_truth[1][0]['Ie']=1e-10

sim_config = SimulatorConfig(delta_pix=delta_pix, num_pix=num_pix, supersample=1, kernel=kernel)
src_model = PhysicalModel([sis.SIS()], [sersic.Sersic(use_lstsq=False)], [sersic.Point(use_lstsq=False)])
src_sim = LensSimulator(src_model, sim_config, bs=1)
src_img = src_sim.simulate(src_truth)

truth[0][0]['theta_E'] = theta_E
phys_model = PhysicalModel([sis.SIS()], [sersic.Sersic(use_lstsq=False)], [sersic.Point(use_lstsq=False)])
sim_config.supersample=supersample

lens_sim = LensSimulator(phys_model, sim_config, bs=1)

print('theta_E location in pixels:', theta_E//delta_pix + ctr)

background_rms, exp_time = 0.005, 400

noiseless_img = lens_sim.simulate(truth)
negidx = noiseless_img < 0
noiseless_img = noiseless_img.at[negidx].set(0)
if negidx.sum() > 10: print('WARNING: OVER 10 NEGATIVE PIXELS IN noiseless_img!', negidx.sum())

I_over_t_exp = noiseless_img/exp_time
observed_img = noiseless_img + np.random.normal(0, background_rms, noiseless_img.shape) + \
                np.random.poisson(I_over_t_exp, noiseless_img.shape)

f, axes = plt.subplots(ncols=2, figsize=(10, 4))
axes = axes.ravel()
srcax = axes[0].imshow(src_img) #, norm=mpl.colors.PowerNorm(0.5, vmin=0, vmax=100))
f.colorbar(srcax, ax=axes[0])
axes[1].set_title('supersamp: {:d}'.format(supersample))
obsax = axes[1].imshow(observed_img) #, norm=mpl.colors.PowerNorm(0.5, vmin=0, vmax=100))
f.colorbar(obsax, ax=axes[1])

plt.show()
prob_model = ForwardProbModel(prior, observed_img, background_rms=background_rms, exp_time=exp_time)
model_seq = ModellingSequence(phys_model, prob_model, sim_config)
```

```python
# note here, we didn't use linear inversion so the brigtness, Ie, was fit for too.
# best_params = prob_model.bij.forward(list(map_estimate[np.nanargmax(lps)]))   
        
# extract_params(best_params)
# print('Compared with truth: \n')

mass_pms = ['theta_E', 'e1', 'e2', 'center_x', 'center_y']
shear_pms = ['gamma1', 'gamma2']
# if not using linear inversion -- then Ie is included
lens_light_pms = ['R_sersic', 'n_sersic', 'center_x', 'center_y', 'Ie']
src_light_pms = ['center_x', 'center_y', 'Ie']

def extract_params(best_params, mass_pms, shear_pms, lens_light_pms, src_light_pms):

    mass_params = {}
    shear_params = {}

    # if using linear inversion -- then no Ie
    # light_pms = ['R_sersic', 'n_sersic', 'e1', 'e2','center_x', 'center_y']

    count = 0
    for i in range(len(best_params)):
        if i==0:
            print('Lens mass params ({:d} sets):'.format(len(best_params[i])))
        elif i==1:
            print('Lens light params({:d} sets):'.format(len(best_params[i])))
        elif i==2:
            print('Source light params({:d} sets):'.format(len(best_params[i])))
        for j in range(len(best_params[i])):
            if j > 0:
                print()
            for k in range(len(best_params[i][j])):
                # print('i, j, k', i, j, k)
                if i==0:
                    if j==0:
                        print('{:5s}={:6.3f}  '.format(mass_pms[k], best_params[i][j][mass_pms[k]]), end='')
                        mass_params[mass_pms[k]] = np.array(best_params[i][j][mass_pms[k]]).item()
                    elif j==1:
                        print('{:5s}={:6.3f}  '.format(shear_pms[k], best_params[i][j][shear_pms[k]]), end='')
                        shear_params[shear_pms[k]] = np.array(best_params[i][j][shear_pms[k]]).item()
                elif i==1:
                    print('{:5s}={:6.3f}  '.format(lens_light_pms[k], best_params[i][j][lens_light_pms[k]]), end='')

                elif i==2:
                    print('{:5s}={:6.3f}  '.format(src_light_pms[k], best_params[i][j][src_light_pms[k]]), end='')

                count += 1
        print('\n')
    return [mass_params, shear_params]

print('The truth parameters:')
extract_params(truth, mass_pms, shear_pms, lens_light_pms, src_light_pms)
```
